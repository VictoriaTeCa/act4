```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cufflinks as cf
from IPython.display import display,HTML
```

```python
#Leo mi archivo.csv
Covid = pd.read_csv("COVID19Mexico.csv")
#Filtro por filas
Covid= Covid.iloc[0:100000,: ]
Covid
```

```python
valores_nulos=Covid.isnull().sum()
valores_nulos
```

```python
#contamos con una base de datos que presenta datos nulos, para poder continuar sin los datos faltantes vamos a sacar el promedio de las datos que si están completos y sustituir los faltantes con el promedio. 
Covid["EMBARAZO"]=Covid["EMBARAZO"].fillna(round(Covid["EMBARAZO"].mean(),1))
Covid["EDAD"]=Covid["EDAD"].fillna(round(Covid["EDAD"].mean(),1))
Covid["SECTOR"]=Covid["SECTOR"].fillna(round(Covid["SECTOR"].mean(),1))
Covid["ENTIDAD_UM"]=Covid["ENTIDAD_UM"].fillna(round(Covid["ENTIDAD_UM"].mean(),1))
Covid["SEXO"]=Covid["SEXO"].fillna(round(Covid["SEXO"].mean(),1))
Covid["MUNICIPIO_RES"]=Covid["MUNICIPIO_RES"].fillna(round(Covid["MUNICIPIO_RES"].mean(),1))
```

```python
#comprobamos que ya no tenemos datos nulos
valores_nulos=Covid.isnull().sum()
valores_nulos
```

```python
#quitamos los valores nulos de la fila con datos textuales
Covid= Covid.dropna(0)
valores_nulos=Covid.isnull().sum()
valores_nulos
```

```python
#se decidio hacer los outliers de la variable de edad ya que es la unica que presenta outliers de todas las demás variables a comparar. 
fig = plt.figure(figsize =(5, 3))
plt.boxplot(Covid["EDAD"]) 
plt.title("Outliers de Edad")
plt.show() #dibujamos el diagrama
```
